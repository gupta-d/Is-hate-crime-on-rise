```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

file = "hate_crime_v3.csv"

df = pd.read_csv(file, low_memory = False)
df.info()
```

```python
df = df.loc[df['DATA_YEAR']>=2007, :]
df.info()
```

```python
df.head()
```

```python
df['INCIDENT_DATE'] = pd.to_datetime(df['INCIDENT_DATE'])
df['INCIDENT_DATE'] = pd.DatetimeIndex(df['INCIDENT_DATE']).month
months = df['INCIDENT_DATE'].unique()
months
```

```python
months.sort(kind='quicksort')
months
```

```python
month_names = df["DATA_MONTH"].unique()
month_count = df["DATA_MONTH"].value_counts()
month_count
```

```python
xticks = ["May", "Aug", "Jul", "Oct", "Sep", "Jun", "Apr", "Mar", "Nov", "Jan", "Feb", "Dec"]

plt.bar(month_names, month_count)
plt.title("Total Hate Crimes Each Month")
plt.ylabel("Number of Hate Crimes")
plt.xlabel("Calendar Month")
plt.xticks(month_names, rotation="vertical")
plt.
```

```python
xticks = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

plt.plot(months, month_count)
plt.title("Total Hate Crimes Each Month")
plt.ylabel("Number of Hate Crimes")
plt.xlabel("Calendar Month")
plt.xticks(months, labels=xticks)
plt.grid()
```

```python
year_one = df.loc[df["DATA_YEAR"] == 2007]
year_one_count = year_one["DATA_MONTH"].value_counts()

year_two = df.loc[df["DATA_YEAR"] == 2008]
year_two_count = year_two["DATA_MONTH"].value_counts()

year_three = df.loc[df["DATA_YEAR"] == 2009]
year_three_count = year_three["DATA_MONTH"].value_counts()

year_four = df.loc[df["DATA_YEAR"] == 2010]
year_four_count = year_four["DATA_MONTH"].value_counts()

year_five = df.loc[df["DATA_YEAR"] == 2011]
year_five_count = year_five["DATA_MONTH"].value_counts()

year_six = df.loc[df["DATA_YEAR"] == 2012]
year_six_count = year_six["DATA_MONTH"].value_counts()

year_seven = df.loc[df["DATA_YEAR"] == 2013]
year_seven_count = year_seven["DATA_MONTH"].value_counts()

year_eight = df.loc[df["DATA_YEAR"] == 2014]
year_eight_count = year_eight["DATA_MONTH"].value_counts()

year_nine = df.loc[df["DATA_YEAR"] == 2015]
year_nine_count = year_nine["DATA_MONTH"].value_counts()

year_ten = df.loc[df["DATA_YEAR"] == 2016]
year_ten_count = year_ten["DATA_MONTH"].value_counts()

year_eleven = df.loc[df["DATA_YEAR"] == 2017]
year_eleven_count = year_eleven["DATA_MONTH"].value_counts()
```

```python
month_names = df["DATA_MONTH"].unique()
month_names.sort(kind='quicksort')
month_names
```

```python
summary_df = pd.DataFrame({"Month": month_names,
                           "2007": year_one_count,
                          "2008": year_two_count,
                          "2009": year_three_count,
                          "2010": year_four_count,
                          "2011": year_five_count,
                          "2012": year_six_count,
                          "2013": year_seven_count,
                          "2014": year_eight_count,
                          "2015": year_nine_count,
                          "2016": year_ten_count,
                          "2017": year_eleven_count})
summary_df
```

```python
plt.plot(month_names, summary_df["2007"])
plt.plot(month_names, summary_df["2008"])
plt.plot(month_names, summary_df["2009"])
plt.plot(month_names, summary_df["2010"])
plt.plot(month_names, summary_df["2011"])
plt.plot(month_names, summary_df["2012"])
plt.xticks(month_names, rotation="vertical")
plt.grid()
plt.legend(loc=4)
plt.title("Monthly Flucuation by Year")
plt.ylabel("Number of Hate Crimes")
plt.xlabel("Calendar Month")
```

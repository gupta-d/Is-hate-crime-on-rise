```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

file = "../input_files/hate_crime_v3.csv"

df = pd.read_csv(file, low_memory = False)
df.info()
```

```python
df = df.loc[df['DATA_YEAR']>=2007, :]
df.info()
```

```python
df.head()
```

```python
df['INCIDENT_DATE'] = pd.to_datetime(df['INCIDENT_DATE'])
df['INCIDENT_DATE'] = pd.DatetimeIndex(df['INCIDENT_DATE']).month
months = df['INCIDENT_DATE'].unique()
```

```python
months
```

```python
months.sort(kind='quicksort')
months
```

```python
month_count = df["INCIDENT_DATE"].value_counts()
months_df = pd.DataFrame({"Total": month_count})
months_df = months_df.sort_index()
months_df
```

```python
xticks = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

plt.plot(months, months_df.Total)
plt.title("Total Hate Crimes Each Month")
plt.ylabel("Number of Hate Crimes")
plt.xlabel("Calendar Month")
plt.xticks(months, labels=xticks)
plt.grid()
plt.savefig('Images/Hate_Crimes_by_Month_v2.png')
```

```python
xticks = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

plt.plot(months, months_df.Total)
plt.title("Total Hate Crimes Each Month")
plt.ylabel("Number of Hate Crimes")
plt.xlabel("Calendar Month")
plt.xticks(months, labels=xticks)
plt.grid()
plt.ylim(0,8000)
plt.savefig('Images/Hate_Crimes_by_Month.png')
```

```python
year_one = df.loc[df["DATA_YEAR"] == 2007]
year_one_count = year_one["INCIDENT_DATE"].value_counts()

year_two = df.loc[df["DATA_YEAR"] == 2008]
year_two_count = year_two["INCIDENT_DATE"].value_counts()

year_three = df.loc[df["DATA_YEAR"] == 2009]
year_three_count = year_three["INCIDENT_DATE"].value_counts()

year_four = df.loc[df["DATA_YEAR"] == 2010]
year_four_count = year_four["INCIDENT_DATE"].value_counts()

year_five = df.loc[df["DATA_YEAR"] == 2011]
year_five_count = year_five["INCIDENT_DATE"].value_counts()

year_six = df.loc[df["DATA_YEAR"] == 2012]
year_six_count = year_six["INCIDENT_DATE"].value_counts()

year_seven = df.loc[df["DATA_YEAR"] == 2013]
year_seven_count = year_seven["INCIDENT_DATE"].value_counts()

year_eight = df.loc[df["DATA_YEAR"] == 2014]
year_eight_count = year_eight["INCIDENT_DATE"].value_counts()

year_nine = df.loc[df["DATA_YEAR"] == 2015]
year_nine_count = year_nine["INCIDENT_DATE"].value_counts()

year_ten = df.loc[df["DATA_YEAR"] == 2016]
year_ten_count = year_ten["INCIDENT_DATE"].value_counts()

year_eleven = df.loc[df["DATA_YEAR"] == 2017]
year_eleven_count = year_eleven["INCIDENT_DATE"].value_counts()
```

```python
summary_df = pd.DataFrame({"2007": year_one_count,
                          "2008": year_two_count,
                          "2009": year_three_count,
                          "2010": year_four_count,
                          "2011": year_five_count,
                          "2012": year_six_count,
                          "2013": year_seven_count,
                          "2014": year_eight_count,
                          "2015": year_nine_count,
                          "2016": year_ten_count,
                          "2017": year_eleven_count})
summary_df = summary_df.sort_index()
summary_df
```

```python
xticks = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

plt.plot(months, summary_df["2007"])
plt.plot(months, summary_df["2008"])
plt.plot(months, summary_df["2009"])
plt.plot(months, summary_df["2010"])
plt.plot(months, summary_df["2011"])
plt.plot(months, summary_df["2012"])
plt.xticks(months, labels=xticks, rotation="horizontal")
plt.grid()
plt.legend(loc=2)
plt.title("Monthly Flucuation by Year: '07 - '12")
plt.ylabel("Number of Hate Crimes")
plt.xlabel("Calendar Month")
plt.savefig('Images/Month_and_Year_Breakdown_07_to_12.png')
```
